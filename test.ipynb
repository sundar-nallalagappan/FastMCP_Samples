```python
import time

def handle_user(user_id):
    print(f"User {user_id}: request started")
    time.sleep(2)   # blocking wait
    print(f"User {user_id}: response sent")

def main():
    handle_user(1)
    handle_user(2)
    handle_user(3)

main()
```

```python
import time
import asyncio

async def handle_user(user_id):
    print(f"User {user_id}: request started")
    await asyncio.sleep(2)    # blocking wait
    print(f"User {user_id}: response sent")

async def main():
    tasks = [
    handle_user(1),
    handle_user(2),
    handle_user(3)
    ]
    await asyncio.gather(*tasks)
    
await main()
```
